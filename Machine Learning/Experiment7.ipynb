```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Load a sample dataset
data = load_iris()
X, y = data.data, data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 1. Feature Extraction using Principal Component Analysis (PCA)
print("Feature Extraction using PCA:")
n_components_pca = 2  # Number of principal components
pca = PCA(n_components=n_components_pca)
X_train_pca = pca.fit_transform(X_train)
print("PCA Components Shape:", X_train_pca.shape)
```

```python
# 2. Feature Extraction using Linear Discriminant Analysis (LDA)
# Note: LDA requires the target variable (y) for supervised learning
print("\nFeature Extraction using LDA:")
n_components_lda = 2  # Number of components for LDA
lda = LDA(n_components=n_components_lda)
X_train_lda = lda.fit_transform(X_train, y_train)
print("LDA Components Shape:", X_train_lda.shape)
```

```python
# Additional: To visualize the results (if running in a Jupyter Notebook or similar)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=40)
plt.title("PCA Result")

# Plot LDA results
plt.subplot(1, 3, 2)
plt.scatter(X_train_lda[:, 0], X_train_lda[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=40)
plt.title("LDA Result")


plt.show()
```
